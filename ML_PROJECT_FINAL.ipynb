```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import pickle
import os
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define all target variables and their classes
target_variables = {
    'knee_flexion': ['normal', 'under-flexion', 'over-flexion'],
    'saddle_height': ['normal', 'too low', 'too high'],
    'saddle_position': ['normal', 'too forward', 'too back'],
    'reach_label': ['normal', 'overreach', 'underreach'],
    'hip_rocking': ['stable', 'unstable'],
    'hip_asymmetry': ['normal', 'left-lean', 'right-lean'],
    'spine_asymmetry': ['normal', 'left lean', 'right lean']
}

# Path to your CSV file
csv_path = '/content/cycling_posture_dataset_realistic.csv.xls'  # Update with your actual path

# Load data
print("Loading dataset...")
data = pd.read_csv(csv_path)
print(f"Dataset loaded with {data.shape[0]} samples and {data.shape[1]} features")

# Extract feature columns (all columns except target variables)
feature_columns = [col for col in data.columns if col not in target_variables.keys()]
print(f"Input features: {len(feature_columns)}")
print(f"Target variables: {len(target_variables)}")

# Hyperparameters
learning_rate = 1e-3
batch_size = 32  # Reduced from 64 to handle smaller dataset better
epochs = 200     # Reduced from 500 to avoid overfitting
l2_reg = 0.01
early_stopping_patience = 20
lr_reduction_patience = 10
validation_split = 0.2

# Directory to save models
save_dir = "cyclist_dysfunction_models"
os.makedirs(save_dir, exist_ok=True)

# Create label encoders for each target variable
label_encoders = {}
for target in target_variables.keys():
    le = LabelEncoder()
    le.fit(target_variables[target])
    label_encoders[target] = le

# Function to create a model for binary or multi-class classification
def create_model(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # Reduce network size for smaller dataset
    x = Dense(128, activation='relu', kernel_regularizer=l2(l2_reg))(inputs)
    x = Dropout(0.3)(x)
    x = Dense(64, activation='relu', kernel_regularizer=l2(l2_reg))(x)
    x = Dropout(0.3)(x)
    x = Dense(32, activation='relu', kernel_regularizer=l2(l2_reg))(x)

    if num_classes == 2:
        # Binary classification
        outputs = Dense(1, activation='sigmoid')(x)
        loss = 'binary_crossentropy'
    else:
        # Multi-class classification
        outputs = Dense(num_classes, activation='softmax')(x)
        loss = 'sparse_categorical_crossentropy'

    model = Model(inputs, outputs)
    model.compile(
        optimizer=Adam(learning_rate),
        loss=loss,
        metrics=['accuracy']
    )
    return model

# Define callbacks for training
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=early_stopping_patience,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=lr_reduction_patience,
        verbose=1
    )
]

# Train and save each model
for target in target_variables.keys():
    print(f"\n{'='*50}")
    print(f"Training model for: {target}")
    print(f"Classes: {target_variables[target]}")
    print(f"{'='*50}")

    # Prepare data
    X = data[feature_columns].values

    # Transform target variable to numeric using the label encoder
    y_encoded = label_encoders[target].transform(data[target].values)
    num_classes = len(target_variables[target])

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_encoded, test_size=validation_split, random_state=42
    )

    print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
    print(f"Number of classes: {num_classes}")

    # Create and train model
    input_shape = (X_train.shape[1],)
    model = create_model(input_shape, num_classes)

    # Print model summary
    model.summary()

    # Train the model
    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_test, y_test),
        callbacks=callbacks,
        verbose=1
    )

    # Evaluate the model
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"Test accuracy: {test_acc:.4f}")

    # Save model in H5 format
    model_path = os.path.join(save_dir, f"{target}_model.h5")
    model.save(model_path)

    # Save label encoder
    encoder_path = os.path.join(save_dir, f"{target}_encoder.pkl")
    with open(encoder_path, 'wb') as f:
        pickle.dump(label_encoders[target], f)

    # Calculate and display model size in MB
    model_size = os.path.getsize(model_path) / (1024 * 1024)
    print(f"Model size: {model_size:.2f} MB")

    # Plot training history
    plt.figure(figsize=(12, 5))

    # Accuracy Plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{target} Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Loss Plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{target} Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"{target}_training_history.png"))
    plt.show()

print("\nTraining complete! All models saved to:", save_dir)

```

# **SVM**

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
import warnings
warnings.filterwarnings('ignore')  # Suppress warnings

# Define all target variables and their classes
target_variables = {
    'knee_flexion': ['normal', 'under-flexion', 'over-flexion'],
    'saddle_height': ['normal', 'too low', 'too high'],
    'saddle_position': ['normal', 'too forward', 'too back'],
    'reach_label': ['normal', 'overreach', 'underreach'],
    'hip_rocking': ['stable', 'unstable'],
    'hip_asymmetry': ['normal', 'left-lean', 'right-lean'],
    'spine_asymmetry': ['normal', 'left lean', 'right lean']
}

# Path to your CSV file
csv_path = '/content/cycling_posture_dataset_realistic.csv.xls'  # Update with your actual path

# Load data
print("Loading dataset...")
data = pd.read_csv(csv_path)
print(f"Dataset loaded with {data.shape[0]} samples and {data.shape[1]} features")

# Extract feature columns (all columns except target variables)
feature_columns = [col for col in data.columns if col not in target_variables.keys()]
print(f"Input features: {len(feature_columns)}")
print(f"Target variables: {len(target_variables)}")

# Directory to save models
save_dir = "cyclist_dysfunction_svm_models"
os.makedirs(save_dir, exist_ok=True)

# Create label encoders for each target variable
label_encoders = {}
for target in target_variables.keys():
    le = LabelEncoder()
    le.fit(target_variables[target])
    label_encoders[target] = le

# Function to check if all classes are present in a dataset
def check_class_balance(y, target_name):
    """
    Checks if all classes are present in the dataset and returns class distribution

    Args:
        y: Target labels
        target_name: Name of the target variable

    Returns:
        dict: Class distribution
    """
    unique_classes = np.unique(y)
    expected_classes = len(target_variables[target_name])

    class_counts = {target_variables[target_name][i]: np.sum(y == i) for i in unique_classes}

    # Check if all classes are present
    if len(unique_classes) < expected_classes:
        print(f"WARNING: Not all classes are present in {target_name}. Found {len(unique_classes)} out of {expected_classes} classes.")
        print(f"Class distribution: {class_counts}")
        missing_classes = set(range(expected_classes)) - set(unique_classes)
        print(f"Missing class indices: {missing_classes}")
        print(f"Missing classes: {[target_variables[target_name][i] for i in missing_classes]}")

    return class_counts

# Function to train SVM model for a specific target
def train_svm_for_target(target, X, y, test_size=0.2, random_state=42):
    """
    Trains an SVM model for a specific target variable

    Args:
        target: Target variable name
        X: Feature data
        y: Target data (encoded)
        test_size: Test split ratio
        random_state: Random seed

    Returns:
        tuple: Trained pipeline, test data, test accuracy
    """
    # Check class balance
    print(f"\nChecking class balance for {target}...")
    class_distribution = check_class_balance(y, target)
    print(f"Class distribution: {class_distribution}")

    # Make sure we have at least 2 classes in the dataset
    unique_classes = np.unique(y)
    if len(unique_classes) < 2:
        print(f"ERROR: {target} has only {len(unique_classes)} class in the dataset. SVM requires at least 2 classes.")
        return None, None, None, None, None

    try:
        # Use stratified split to maintain class distribution
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, stratify=y, random_state=random_state
        )

        # Check if all classes are in the training set
        train_classes = np.unique(y_train)
        if len(train_classes) < len(unique_classes):
            print(f"WARNING: Training set for {target} has fewer classes ({len(train_classes)}) than the full dataset ({len(unique_classes)}).")
            print("Using stratified k-fold cross-validation instead...")

            # Use stratified k-fold to ensure all classes are represented
            skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)
            for train_index, test_index in skf.split(X, y):
                X_train, X_test = X[train_index], X[test_index]
                y_train, y_test = y[train_index], y[test_index]
                train_classes = np.unique(y_train)
                if len(train_classes) == len(unique_classes):
                    print("Found a split with all classes represented.")
                    break

        # Create and train the SVM pipeline
        print(f"Training SVM for {target}...")
        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('svm', SVC(kernel='rbf', C=1.0, gamma='scale', probability=True, random_state=random_state))
        ])

        # Train the model
        pipeline.fit(X_train, y_train)

        # Evaluate on test set
        y_pred = pipeline.predict(X_test)
        accuracy = np.mean(y_pred == y_test)
        print(f"Test accuracy: {accuracy:.4f}")

        return pipeline, X_test, y_test, y_pred, accuracy

    except Exception as e:
        print(f"Error training SVM for {target}: {str(e)}")
        return None, None, None, None, None

# Train SVM for each target variable
results = {}
for target in target_variables.keys():
    print(f"\n{'='*50}")
    print(f"Processing: {target}")
    print(f"{'='*50}")

    # Prepare data
    X = data[feature_columns].values

    # Transform target variable to numeric using the label encoder
    y_encoded = label_encoders[target].transform(data[target].values)

    # Train SVM model
    pipeline, X_test, y_test, y_pred, accuracy = train_svm_for_target(target, X, y_encoded)

    if pipeline is not None:
        # Save the model
        model_path = os.path.join(save_dir, f"{target}_svm_model.pkl")
        with open(model_path, 'wb') as f:
            pickle.dump(pipeline, f)

        # Save label encoder
        encoder_path = os.path.join(save_dir, f"{target}_encoder.pkl")
        with open(encoder_path, 'wb') as f:
            pickle.dump(label_encoders[target], f)

        # Generate classification report
        print("\nClassification Report:")
        class_names = [target_variables[target][i] for i in range(len(target_variables[target]))]
        report = classification_report(y_test, y_pred, target_names=class_names)
        print(report)

        # Save report to file
        with open(os.path.join(save_dir, f"{target}_classification_report.txt"), 'w') as f:
            f.write(report)

        # Plot confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=class_names,
                    yticklabels=class_names)
        plt.title(f'Confusion Matrix for {target} (SVM)')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.tight_layout()
        plt.savefig(os.path.join(save_dir, f"{target}_confusion_matrix.png"))
        plt.close()

        # Store results
        results[target] = {
            'accuracy': accuracy,
            'model_path': model_path
        }
    else:
        print(f"Skipping {target} due to errors or insufficient classes.")

# Print summary of results
print("\n" + "="*60)
print("Summary of SVM Model Results")
print("="*60)
for target, result in results.items():
    print(f"{target}: Accuracy = {result['accuracy']:.4f}, Model saved at {result['model_path']}")
print("\nTraining complete! All models saved to:", save_dir)

# Function to make predictions for a new cyclist
def predict_for_cyclist(cyclist_data):
    """
    Makes predictions for a single cyclist data point using SVM models

    Args:
        cyclist_data: Dictionary or Series with feature values

    Returns:
        Dictionary with predictions for each target variable
    """
    # Convert input to DataFrame if it's a dictionary
    if isinstance(cyclist_data, dict):
        cyclist_data = pd.DataFrame([cyclist_data])
    elif isinstance(cyclist_data, pd.Series):
        cyclist_data = pd.DataFrame([cyclist_data])

    # Ensure we have all required features
    missing_features = [col for col in feature_columns if col not in cyclist_data.columns]
    if missing_features:
        raise ValueError(f"Missing required features: {missing_features}")

    # Extract just the features we need in the right order
    input_data = cyclist_data[feature_columns].values

    # Make predictions for each target variable
    results = {}
    for target in target_variables.keys():
        try:
            # Load model
            model_path = os.path.join(save_dir, f"{target}_svm_model.pkl")
            if not os.path.exists(model_path):
                print(f"Model for {target} not found, skipping...")
                continue

            with open(model_path, 'rb') as f:
                pipeline = pickle.load(f)

            # Load label encoder
            encoder_path = os.path.join(save_dir, f"{target}_encoder.pkl")
            with open(encoder_path, 'rb') as f:
                encoder = pickle.load(f)

            # Make prediction
            pred_probs = pipeline.predict_proba(input_data)[0]
            pred_class = pipeline.predict(input_data)[0]

            # Convert numeric prediction back to label
            pred_label = encoder.inverse_transform([pred_class])[0]

            # Get confidence (probability of predicted class)
            confidence = pred_probs[pred_class]

            # Store result
            results[target] = {
                'prediction': pred_label,
                'confidence': float(confidence)
            }
        except Exception as e:
            print(f"Error predicting {target}: {str(e)}")

    return results

# Example: Try prediction if there's data available
if len(data) > 0:
    print("\nExample prediction for a sample cyclist:")
    sample_cyclist = data.iloc[0][feature_columns]
    try:
        predictions = predict_for_cyclist(sample_cyclist)
        for target, result in predictions.items():
            print(f"{target}: {result['prediction']} (confidence: {result['confidence']:.2f})")
    except Exception as e:
        print(f"Error making prediction: {str(e)}")
```

# **RANDOM FOREST**

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
import warnings
warnings.filterwarnings('ignore')  # Suppress warnings

# Define all target variables and their classes
target_variables = {
    'knee_flexion': ['normal', 'under-flexion', 'over-flexion'],
    'saddle_height': ['normal', 'too low', 'too high'],
    'saddle_position': ['normal', 'too forward', 'too back'],
    'reach_label': ['normal', 'overreach', 'underreach'],
    'hip_rocking': ['stable', 'unstable'],
    'hip_asymmetry': ['normal', 'left-lean', 'right-lean'],
    'spine_asymmetry': ['normal', 'left lean', 'right lean']
}

# Path to your CSV file
csv_path = '/content/cycling_posture_dataset_realistic.csv.xls'  # Update with your actual path

# Load data
print("Loading dataset...")
data = pd.read_csv(csv_path)
print(f"Dataset loaded with {data.shape[0]} samples and {data.shape[1]} features")

# Extract feature columns (all columns except target variables)
feature_columns = [col for col in data.columns if col not in target_variables.keys()]
print(f"Input features: {len(feature_columns)}")
print(f"Target variables: {len(target_variables)}")

# Directory to save models
save_dir = "cyclist_dysfunction_rf_models"
os.makedirs(save_dir, exist_ok=True)

# Create label encoders for each target variable
label_encoders = {}
for target in target_variables.keys():
    le = LabelEncoder()
    le.fit(target_variables[target])
    label_encoders[target] = le

# Function to check if all classes are present in a dataset
def check_class_balance(y, target_name):
    """
    Checks if all classes are present in the dataset and returns class distribution

    Args:
        y: Target labels
        target_name: Name of the target variable

    Returns:
        dict: Class distribution
    """
    unique_classes = np.unique(y)
    expected_classes = len(target_variables[target_name])

    class_counts = {target_variables[target_name][i]: np.sum(y == i) for i in unique_classes}

    # Check if all classes are present
    if len(unique_classes) < expected_classes:
        print(f"WARNING: Not all classes are present in {target_name}. Found {len(unique_classes)} out of {expected_classes} classes.")
        print(f"Class distribution: {class_counts}")
        missing_classes = set(range(expected_classes)) - set(unique_classes)
        print(f"Missing class indices: {missing_classes}")
        print(f"Missing classes: {[target_variables[target_name][i] for i in missing_classes]}")

    return class_counts

# Train Random Forest for each target variable
results = {}
for target in target_variables.keys():
    print(f"\n{'='*50}")
    print(f"Processing: {target}")
    print(f"{'='*50}")

    # Prepare data
    X = data[feature_columns].values

    # Transform target variable to numeric using the label encoder
    y_encoded = label_encoders[target].transform(data[target].values)

    # Check class balance
    print(f"Checking class balance for {target}...")
    class_distribution = check_class_balance(y_encoded, target)
    print(f"Class distribution: {class_distribution}")

    # Make sure we have at least 2 classes in the dataset (RF can handle it, but it's not useful)
    unique_classes = np.unique(y_encoded)
    if len(unique_classes) < 2:
        print(f"Warning: {target} has only {len(unique_classes)} class in the dataset. Skipping.")
        continue

    try:
        # Use stratified split to maintain class distribution
        X_train, X_test, y_train, y_test = train_test_split(
            X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
        )

        # Check if all classes are in the training set
        train_classes = np.unique(y_train)
        if len(train_classes) < len(unique_classes):
            print(f"WARNING: Training set for {target} has fewer classes ({len(train_classes)}) than the full dataset ({len(unique_classes)}).")
            print("Using stratified k-fold cross-validation instead...")

            # Use stratified k-fold to ensure all classes are represented
            skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
            for train_index, test_index in skf.split(X, y_encoded):
                X_train, X_test = X[train_index], X[test_index]
                y_train, y_test = y_encoded[train_index], y_encoded[test_index]
                train_classes = np.unique(y_train)
                if len(train_classes) == len(unique_classes):
                    print("Found a split with all classes represented.")
                    break

        # Create and train the Random Forest model
        print(f"Training Random Forest for {target}...")
        rf = RandomForestClassifier(
            n_estimators=100,
            max_depth=None,
            min_samples_split=2,
            min_samples_leaf=1,
            max_features='sqrt',
            bootstrap=True,
            n_jobs=-1,
            random_state=42
        )

        # Train the model
        rf.fit(X_train, y_train)

        # Evaluate on test set
        y_pred = rf.predict(X_test)
        accuracy = np.mean(y_pred == y_test)
        print(f"Test accuracy: {accuracy:.4f}")

        # Save the model
        model_path = os.path.join(save_dir, f"{target}_rf_model.pkl")
        with open(model_path, 'wb') as f:
            pickle.dump(rf, f)

        # Save label encoder
        encoder_path = os.path.join(save_dir, f"{target}_encoder.pkl")
        with open(encoder_path, 'wb') as f:
            pickle.dump(label_encoders[target], f)

        # Generate classification report
        print("\nClassification Report:")
        class_names = [target_variables[target][i] for i in range(len(target_variables[target]))]
        report = classification_report(y_test, y_pred, target_names=class_names)
        print(report)

        # Save report to file
        with open(os.path.join(save_dir, f"{target}_classification_report.txt"), 'w') as f:
            f.write(report)

        # Plot confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=class_names,
                    yticklabels=class_names)
        plt.title(f'Confusion Matrix for {target} (Random Forest)')
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.tight_layout()
        plt.savefig(os.path.join(save_dir, f"{target}_confusion_matrix.png"))
        plt.close()

        # Feature importance analysis
        importance = rf.feature_importances_
        feature_importance = pd.DataFrame({
            'Feature': feature_columns,
            'Importance': importance
        }).sort_values('Importance', ascending=False)

        print("\nTop 10 Important Features:")
        print(feature_importance.head(10))

        # Save feature importance to CSV
        feature_importance.to_csv(os.path.join(save_dir, f"{target}_feature_importance.csv"), index=False)

        # Plot feature importance (top 15)
        plt.figure(figsize=(12, 8))
        top_features = feature_importance.head(15)
        sns.barplot(x='Importance', y='Feature', data=top_features)
        plt.title(f'Top 15 Feature Importance for {target}')
        plt.tight_layout()
        plt.savefig(os.path.join(save_dir, f"{target}_feature_importance.png"))
        plt.close()

        # Store results
        results[target] = {
            'accuracy': accuracy,
            'model_path': model_path,
            'top_features': feature_importance.head(5)['Feature'].tolist()
        }
    except Exception as e:
        print(f"Error training Random Forest for {target}: {str(e)}")

# Print summary of results
print("\n" + "="*60)
print("Summary of Random Forest Model Results")
print("="*60)
for target, result in results.items():
    print(f"{target}:")
    print(f"  Accuracy: {result['accuracy']:.4f}")
    print(f"  Model saved at: {result['model_path']}")
    print(f"  Top 5 features: {result['top_features']}")
    print()

# Function to make predictions for a new cyclist using all models
def predict_with_rf(cyclist_data):
    """
    Makes predictions for a single cyclist using Random Forest models

    Args:
        cyclist_data: DataFrame or Series with cyclist measurements

    Returns:
        dict: Predictions for each target variable
    """
    # Convert input to DataFrame if it's a dictionary or Series
    if isinstance(cyclist_data, dict):
        cyclist_data = pd.DataFrame([cyclist_data])
    elif isinstance(cyclist_data, pd.Series):
        cyclist_data = pd.DataFrame([cyclist_data])

    # Ensure we have all required features
    missing_features = [col for col in feature_columns if col not in cyclist_data.columns]
    if missing_features:
        raise ValueError(f"Missing required features: {missing_features}")

    # Extract features in the correct order
    X = cyclist_data[feature_columns].values

    # Make predictions for each target variable
    predictions = {}
    for target in target_variables.keys():
        try:
            # Load model
            model_path = os.path.join(save_dir, f"{target}_rf_model.pkl")
            if not os.path.exists(model_path):
                print(f"Model for {target} not found, skipping...")
                continue

            with open(model_path, 'rb') as f:
                rf_model = pickle.load(f)

            # Load label encoder
            encoder_path = os.path.join(save_dir, f"{target}_encoder.pkl")
            with open(encoder_path, 'rb') as f:
                encoder = pickle.load(f)

            # Make prediction
            pred_class = rf_model.predict(X)[0]
            pred_probs = rf_model.predict_proba(X)[0]

            # Get class label
            pred_label = encoder.inverse_transform([pred_class])[0]

            # Get confidence (probability of predicted class)
            confidence = pred_probs[pred_class]

            # Store result
            predictions[target] = {
                'prediction': pred_label,
                'confidence': float(confidence),
                'all_probabilities': {
                    encoder.inverse_transform([i])[0]: float(prob)
                    for i, prob in enumerate(pred_probs)
                }
            }
        except Exception as e:
            print(f"Error predicting {target}: {str(e)}")

    return predictions

# Create an integrated analysis function to provide recommendations
def analyze_cyclist_form(cyclist_data):
    """
    Analyzes a cyclist's form and provides recommendations

    Args:
        cyclist_data: DataFrame or Series with cyclist measurements

    Returns:
        dict: Analysis and recommendations
    """
    # Get predictions
    predictions = predict_with_rf(cyclist_data)

    # Generate analysis
    analysis = {
        'predictions': predictions,
        'issues_detected': [],
        'recommendations': []
    }

    # Check for issues and provide recommendations
    if 'knee_flexion' in predictions:
        pred = predictions['knee_flexion']['prediction']
        if pred == 'under-flexion':
            analysis['issues_detected'].append('Under-flexion of the knee detected')
            analysis['recommendations'].append('Consider increasing saddle height slightly')
        elif pred == 'over-flexion':
            analysis['issues_detected'].append('Over-flexion of the knee detected')
            analysis['recommendations'].append('Consider lowering saddle height slightly')

    if 'saddle_height' in predictions:
        pred = predictions['saddle_height']['prediction']
        if pred == 'too low':
            analysis['issues_detected'].append('Saddle height too low')
            analysis['recommendations'].append('Raise saddle to achieve 70-75° minimum knee angle')
        elif pred == 'too high':
            analysis['issues_detected'].append('Saddle height too high')
            analysis['recommendations'].append('Lower saddle to achieve 70-75° minimum knee angle')

    if 'saddle_position' in predictions:
        pred = predictions['saddle_position']['prediction']
        if pred == 'too forward':
            analysis['issues_detected'].append('Saddle positioned too far forward')
            analysis['recommendations'].append('Move saddle back to achieve proper knee-to-pedal alignment')
        elif pred == 'too back':
            analysis['issues_detected'].append('Saddle positioned too far back')
            analysis['recommendations'].append('Move saddle forward to achieve proper knee-to-pedal alignment')

    if 'reach_label' in predictions:
        pred = predictions['reach_label']['prediction']
        if pred == 'overreach':
            analysis['issues_detected'].append('Overreaching detected')
            analysis['recommendations'].append('Consider a shorter stem or more upright position')
        elif pred == 'underreach':
            analysis['issues_detected'].append('Underreaching detected')
            analysis['recommendations'].append('Consider a longer stem or more stretched position')

    if 'hip_rocking' in predictions:
        pred = predictions['hip_rocking']['prediction']
        if pred == 'unstable':
            analysis['issues_detected'].append('Hip rocking/instability detected')
            analysis['recommendations'].append('Check saddle height and work on core stability')

    if 'hip_asymmetry' in predictions:
        pred = predictions['hip_asymmetry']['prediction']
        if pred != 'normal':
            analysis['issues_detected'].append(f'Hip asymmetry detected: {pred}')
            analysis['recommendations'].append('Consider a professional bike fit to address asymmetry')

    if 'spine_asymmetry' in predictions:
        pred = predictions['spine_asymmetry']['prediction']
        if pred != 'normal':
            analysis['issues_detected'].append(f'Spine asymmetry detected: {pred}')
            analysis['recommendations'].append('Consider a professional bike fit and possibly physical therapy')

    # Add general recommendation if there are issues
    if analysis['issues_detected']:
        analysis['recommendations'].append('Consider a comprehensive professional bike fit for optimal riding position')
    else:
        analysis['recommendations'].append('Your cycling form appears good based on the current measurements')

    return analysis

# Example: Try prediction if there's data available
if len(data) > 0:
    print("\nExample analysis for a sample cyclist:")
    sample_cyclist = data.iloc[0][feature_columns]
    try:
        analysis = analyze_cyclist_form(sample_cyclist)

        print("\nPredicted cycling form issues:")
        for issue in analysis['issues_detected']:
            print(f"- {issue}")

        print("\nRecommendations:")
        for recommendation in analysis['recommendations']:
            print(f"- {recommendation}")

        print("\nDetailed predictions:")
        for target, result in analysis['predictions'].items():
            print(f"{target}: {result['prediction']} (confidence: {result['confidence']:.2f})")
    except Exception as e:
        print(f"Error analyzing cyclist form: {str(e)}")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import pickle
import os
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define all target variables and their classes
target_variables = {
    'knee_flexion': ['normal', 'under-flexion', 'over-flexion'],
    'saddle_height': ['normal', 'too low', 'too high'],
    'saddle_position': ['normal', 'too forward', 'too back'],
    'reach_label': ['normal', 'overreach', 'underreach'],
    'hip_rocking': ['stable', 'unstable'],
    'hip_asymmetry': ['normal', 'left-lean', 'right-lean'],
    'spine_asymmetry': ['normal', 'left lean', 'right lean']
}

# Map each target variable to its relevant features
feature_mapping = {
    'knee_flexion': ['min_knee_angle', 'max_knee_angle'],
    'saddle_height': ['min_knee_angle'],
    'saddle_position': ['horizontal_distance', 'rider_inseam_length'],
    'reach_label': ['reach', 'rider_arm_length', 'rider_torso_length', 'horizontal_distance', 'rider_inseam_length'],
    'hip_rocking': ['pelvis_angle_deviation'],
    'hip_asymmetry': ['mean_pelvis_angle'],
    'spine_asymmetry': ['mean_spine_angle']
}

# Path to your CSV file
csv_path = '/content/cycling_posture_dataset_v4.csv.xls'  # Update with your actual path

# Load data
print("Loading dataset...")
data = pd.read_csv(csv_path)
print(f"Dataset loaded with {data.shape[0]} samples and {data.shape[1]} features")

# Display available columns for verification
print("Available columns in dataset:")
print(data.columns.tolist())

# Hyperparameters
learning_rate = 1e-3
batch_size = 32  # Reduced from 64 to handle smaller dataset better
epochs = 200     # Reduced from 500 to avoid overfitting
l2_reg = 0.01
early_stopping_patience = 20
lr_reduction_patience = 10
validation_split = 0.2

# Directory to save models
save_dir = "cyclist_dysfunction_models"
os.makedirs(save_dir, exist_ok=True)

# Create label encoders for each target variable
label_encoders = {}
for target in target_variables.keys():
    le = LabelEncoder()
    le.fit(target_variables[target])
    label_encoders[target] = le

# Function to create a model for binary or multi-class classification
def create_model(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # Reduce network size for smaller dataset
    x = Dense(128, activation='relu', kernel_regularizer=l2(l2_reg))(inputs)
    x = Dropout(0.3)(x)
    x = Dense(64, activation='relu', kernel_regularizer=l2(l2_reg))(x)
    x = Dropout(0.3)(x)
    x = Dense(32, activation='relu', kernel_regularizer=l2(l2_reg))(x)

    if num_classes == 2:
        # Binary classification
        outputs = Dense(1, activation='sigmoid')(x)
        loss = 'binary_crossentropy'
    else:
        # Multi-class classification
        outputs = Dense(num_classes, activation='softmax')(x)
        loss = 'sparse_categorical_crossentropy'

    model = Model(inputs, outputs)
    model.compile(
        optimizer=Adam(learning_rate),
        loss=loss,
        metrics=['accuracy']
    )
    return model

# Define callbacks for training
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=early_stopping_patience,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=lr_reduction_patience,
        verbose=1
    )
]

# Function to train a model
def train_model(target, features, data, label_encoders, save_dir):
    print(f"\n{'='*50}")
    print(f"Training model for: {target}")
    print(f"Selected features: {features}")
    print(f"Classes: {target_variables[target]}")
    print(f"{'='*50}")

    # Verify that all requested features exist in the dataset
    missing_features = [feature for feature in features if feature not in data.columns]
    if missing_features:
        print(f"ERROR: The following features are missing from the dataset: {missing_features}")
        print("Available columns:", data.columns.tolist())
        return

    # Prepare data with only the selected features
    X = data[features].values

    # Handle missing values if any
    if np.isnan(X).any():
        print("Warning: Dataset contains missing values. Filling with zeros.")
        X = np.nan_to_num(X, nan=0.0)

    # Transform target variable to numeric using the label encoder
    y_encoded = label_encoders[target].transform(data[target].values)
    num_classes = len(target_variables[target])

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_encoded, test_size=validation_split, random_state=42, stratify=y_encoded
    )

    print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
    print(f"Number of classes: {num_classes}")
    print(f"Feature input shape: {X_train.shape[1]}")

    # Create and train model
    input_shape = (X_train.shape[1],)
    model = create_model(input_shape, num_classes)

    # Print model summary
    model.summary()

    # Train the model
    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_test, y_test),
        callbacks=callbacks,
        verbose=1
    )

    # Evaluate the model
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"Test accuracy: {test_acc:.4f}")

    # Save model in H5 format
    model_path = os.path.join(save_dir, f"{target}_model.h5")
    model.save(model_path)

    # Save label encoder
    encoder_path = os.path.join(save_dir, f"{target}_encoder.pkl")
    with open(encoder_path, 'wb') as f:
        pickle.dump(label_encoders[target], f)

    # Calculate and display model size in MB
    model_size = os.path.getsize(model_path) / (1024 * 1024)
    print(f"Model size: {model_size:.2f} MB")

    # Save a file with the features used
    features_path = os.path.join(save_dir, f"{target}_features.txt")
    with open(features_path, 'w') as f:
        f.write("\n".join(features))

    # Plot training history
    plt.figure(figsize=(12, 5))

    # Accuracy Plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{target} Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Loss Plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{target} Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"{target}_training_history.png"))
    plt.show()

    return model, history

# Uncomment the model you want to train below
# You can also train multiple models by uncommenting multiple sections

# ============ KNEE FLEXION MODEL ============
train_model(
    target='knee_flexion',
    features=feature_mapping['knee_flexion'],
    data=data,
    label_encoders=label_encoders,
    save_dir=save_dir
)

print("\nTraining complete! All models saved to:", save_dir)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import pickle
import os
import time
import seaborn as sns

# Define all target variables and their classes
target_variables = {
    'knee_flexion': ['normal', 'under-flexion', 'over-flexion'],
    'saddle_height': ['normal', 'too low', 'too high'],
    'saddle_position': ['normal', 'too forward', 'too back'],
    'reach_label': ['normal', 'overreach', 'underreach'],
    'hip_rocking': ['stable', 'unstable'],
    'hip_asymmetry': ['normal', 'left-lean', 'right-lean'],
    'spine_asymmetry': ['normal', 'left lean', 'right lean']
}

# Map each target variable to its relevant features
feature_mapping = {
    'knee_flexion': ['min_knee_angle', 'max_knee_angle'],
    'saddle_height': ['min_knee_angle','max_knee_angle'],
    'saddle_position': ['horizontal_distance', 'rider_inseam_length'],
    'reach_label': ['reach', 'rider_arm_length', 'rider_torso_length', 'horizontal_distance', 'rider_inseam_length'],
    'hip_rocking': ['pelvis_angle_deviation'],
    'hip_asymmetry': ['mean_pelvis_angle'],
    'spine_asymmetry': ['mean_spine_angle']
}

# Path to your CSV file
csv_path = '/content/cycling_posture_dataset_v4.csv.xls'  # Update with your actual path

# Load data
print("Loading dataset...")
data = pd.read_csv(csv_path)
print(f"Dataset loaded with {data.shape[0]} samples and {data.shape[1]} features")

# Display available columns for verification
print("Available columns in dataset:")
print(data.columns.tolist())

# Directory to save models
save_dir = "svm_models"
os.makedirs(save_dir, exist_ok=True)

# Create label encoders for each target variable
label_encoders = {}
for target in target_variables.keys():
    le = LabelEncoder()
    le.fit(target_variables[target])
    label_encoders[target] = le

# Define hyperparameters for model training
validation_split = 0.2

# SVM hyperparameters
svm_param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto', 0.1, 0.01],
    'kernel': ['rbf']  # You can add more kernels if needed: 'linear', 'poly', 'sigmoid'
}

# Random Forest hyperparameters
rf_param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

# Dictionary to store both model types
models = {
    'svm': {
        'model': SVC,
        'params': svm_param_grid,
        'name': 'Support Vector Machine'
    },
    'rf': {
        'model': RandomForestClassifier,
        'params': rf_param_grid,
        'name': 'Random Forest'
    }
}

# Function to train a model using either SVM or RF
def train_model(target, features, data, label_encoders, save_dir, model_type='svm'):
    print(f"\n{'='*50}")
    print(f"Training {models[model_type]['name']} model for: {target}")
    print(f"Selected features: {features}")
    print(f"Classes: {target_variables[target]}")
    print(f"{'='*50}")

    # Start timing
    start_time = time.time()

    # Verify that all requested features exist in the dataset
    missing_features = [feature for feature in features if feature not in data.columns]
    if missing_features:
        print(f"ERROR: The following features are missing from the dataset: {missing_features}")
        print("Available columns:", data.columns.tolist())
        return

    # Prepare data with only the selected features
    X = data[features].values

    # Handle missing values if any
    if np.isnan(X).any():
        print("Warning: Dataset contains missing values. Filling with zeros.")
        X = np.nan_to_num(X, nan=0.0)

    # Feature scaling for SVM (important!)
    if model_type == 'svm':
        scaler = StandardScaler()
        X = scaler.fit_transform(X)
        # Save scaler for future use
        scaler_path = os.path.join(save_dir, f"{target}_{model_type}_scaler.pkl")
        with open(scaler_path, 'wb') as f:
            pickle.dump(scaler, f)

    # Transform target variable to numeric using the label encoder
    y_encoded = label_encoders[target].transform(data[target].values)
    num_classes = len(target_variables[target])

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_encoded, test_size=validation_split, random_state=42, stratify=y_encoded
    )

    print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
    print(f"Number of classes: {num_classes}")
    print(f"Feature input shape: {X_train.shape[1]}")

    # Create and train model using GridSearchCV for hyperparameter tuning
    print(f"Performing grid search for optimal hyperparameters...")

    # Initialize the model with grid search
    grid_search = GridSearchCV(
        models[model_type]['model'](),
        models[model_type]['params'],
        cv=5,
        n_jobs=-1,
        verbose=1
    )

    # Fit model
    grid_search.fit(X_train, y_train)

    # Get best model
    best_model = grid_search.best_estimator_
    print(f"Best parameters: {grid_search.best_params_}")

    # Make predictions on test set
    y_pred = best_model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Test accuracy: {accuracy:.4f}")

    # Print detailed classification report
    print("\nClassification Report:")
    class_names = target_variables[target]
    print(classification_report(y_test, y_pred, target_names=class_names))

    # Create confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names,
                yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix - {target} ({models[model_type]["name"]})')
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"{target}_{model_type}_confusion_matrix.png"))

    # Save the trained model
    model_path = os.path.join(save_dir, f"{target}_{model_type}_model.pkl")
    with open(model_path, 'wb') as f:
        pickle.dump(best_model, f)

    # Save label encoder
    encoder_path = os.path.join(save_dir, f"{target}_{model_type}_encoder.pkl")
    with open(encoder_path, 'wb') as f:
        pickle.dump(label_encoders[target], f)

    # Save a file with the features used
    features_path = os.path.join(save_dir, f"{target}_{model_type}_features.txt")
    with open(features_path, 'w') as f:
        f.write("\n".join(features))

    # Save the best parameters
    params_path = os.path.join(save_dir, f"{target}_{model_type}_best_params.txt")
    with open(params_path, 'w') as f:
        for param, value in grid_search.best_params_.items():
            f.write(f"{param}: {value}\n")

    # Calculate and print training time
    training_time = time.time() - start_time
    print(f"Training completed in {training_time:.2f} seconds")

    # If using Random Forest, plot feature importances
    if model_type == 'rf':
        importances = best_model.feature_importances_
        indices = np.argsort(importances)[::-1]

        plt.figure(figsize=(10, 6))
        plt.title(f'Feature Importances for {target}')
        plt.bar(range(X.shape[1]), importances[indices], align='center')
        plt.xticks(range(X.shape[1]), [features[i] for i in indices], rotation=90)
        plt.tight_layout()
        plt.savefig(os.path.join(save_dir, f"{target}_{model_type}_feature_importance.png"))

    return best_model

# Uncomment the model and algorithm you want to train below
# SVM MODELS

# ============ KNEE FLEXION MODEL (SVM) ============
# train_model(
#     target='knee_flexion',
#     features=feature_mapping['knee_flexion'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='svm'
# )

# ============ SADDLE HEIGHT MODEL (SVM) ============
# train_model(
#     target='saddle_height',
#     features=feature_mapping['saddle_height'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='svm'
# )

# ============ SADDLE POSITION MODEL (SVM) ============
# train_model(
#     target='saddle_position',
#     features=feature_mapping['saddle_position'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='svm'
# )

# ============ REACH LABEL MODEL (SVM) ============
# train_model(
#     target='reach_label',
#     features=feature_mapping['reach_label'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='svm'
# )

# ============ HIP ROCKING MODEL (SVM) ============
# train_model(
#     target='hip_rocking',
#     features=feature_mapping['hip_rocking'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='svm'
# )

# ============ HIP ASYMMETRY MODEL (SVM) ============
# train_model(
#     target='hip_asymmetry',
#     features=feature_mapping['hip_asymmetry'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='svm'
# )

# ============ SPINE ASYMMETRY MODEL (SVM) ============
# train_model(
#     target='spine_asymmetry',
#     features=feature_mapping['spine_asymmetry'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='svm'
# )

# RANDOM FOREST MODELS

# ============ KNEE FLEXION MODEL (RF) ============
# train_model(
#     target='knee_flexion',
#     features=feature_mapping['knee_flexion'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='rf'
# )

# ============ SADDLE HEIGHT MODEL (RF) ============
# train_model(
#     target='saddle_height',
#     features=feature_mapping['saddle_height'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='rf'
# )

# ============ SADDLE POSITION MODEL (RF) ============
# train_model(
#     target='saddle_position',
#     features=feature_mapping['saddle_position'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='rf'
# )

# ============ REACH LABEL MODEL (RF) ============
# train_model(
#     target='reach_label',
#     features=feature_mapping['reach_label'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='rf'
# )

# ============ HIP ROCKING MODEL (RF) ============
# train_model(
#     target='hip_rocking',
#     features=feature_mapping['hip_rocking'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='rf'
# )

# ============ HIP ASYMMETRY MODEL (RF) ============
# train_model(
#     target='hip_asymmetry',
#     features=feature_mapping['hip_asymmetry'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='rf'
# )

# ============ SPINE ASYMMETRY MODEL (RF) ============
# train_model(
#     target='spine_asymmetry',
#     features=feature_mapping['spine_asymmetry'],
#     data=data,
#     label_encoders=label_encoders,
#     save_dir=save_dir,
#     model_type='rf'
# )

# ============ TRAIN ALL MODELS SEQUENTIALLY WITH BOTH ALGORITHMS ============
# Uncomment this section if you want to train all models with both algorithms
model_types = ['svm', 'rf']
for model_type in model_types:
    print(f"\n\n{'#'*80}")
    print(f"# Training all models using {models[model_type]['name']}")
    print(f"{'#'*80}\n")
    for target, features in feature_mapping.items():
        train_model(
            target=target,
            features=features,
            data=data,
            label_encoders=label_encoders,
            save_dir=save_dir,
            model_type=model_type
        )

print("\nTraining complete! All models saved to:", save_dir)
```

# **RF -> SVM**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import pickle
import os
import time
import seaborn as sns

# Define all target variables and their classes
target_variables = {
    'knee_flexion': ['normal', 'under-flexion', 'over-flexion'],
    'saddle_height': ['normal', 'too low', 'too high'],
    'saddle_position': ['normal', 'too forward', 'too back'],
    'reach_label': ['normal', 'overreach', 'underreach'],
    'hip_rocking': ['stable', 'unstable'],
    'hip_asymmetry': ['normal', 'left-lean', 'right-lean'],
    'spine_asymmetry': ['normal', 'left lean', 'right lean']
}

# Original feature mapping - we'll now use this only as a reference
original_feature_mapping = {
    'knee_flexion': ['min_knee_angle', 'max_knee_angle'],
    'saddle_height': ['min_knee_angle'],
    'saddle_position': ['horizontal_distance', 'rider_inseam_length'],
    'reach_label': ['reach', 'rider_arm_length', 'rider_torso_length', 'horizontal_distance', 'rider_inseam_length'],
    'hip_rocking': ['pelvis_angle_deviation'],
    'hip_asymmetry': ['mean_pelvis_angle'],
    'spine_asymmetry': ['mean_spine_angle']
}

# Path to your CSV file
csv_path = '/content/cycling_posture_dataset_v4.csv.xls'  # Update with your actual path

# Load data
print("Loading dataset...")
data = pd.read_csv(csv_path)
print(f"Dataset loaded with {data.shape[0]} samples and {data.shape[1]} features")

# Display available columns for verification
print("Available columns in dataset:")
print(data.columns.tolist())

# Directory to save models
save_dir = "rf_to_svm_models"
os.makedirs(save_dir, exist_ok=True)

# Directory to save feature importance analysis
feature_dir = os.path.join(save_dir, "feature_analysis")
os.makedirs(feature_dir, exist_ok=True)

# Create label encoders for each target variable
label_encoders = {}
for target in target_variables.keys():
    le = LabelEncoder()
    le.fit(target_variables[target])
    label_encoders[target] = le

# Define hyperparameters for model training
validation_split = 0.2

# Random Forest hyperparameters for feature selection
rf_feature_selection_params = {
    'n_estimators': 100,
    'max_depth': None,
    'min_samples_split': 2,
    'random_state': 42
}

# SVM hyperparameters
svm_param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto', 0.1, 0.01],
    'kernel': ['rbf']  # You can add more kernels if needed: 'linear', 'poly', 'sigmoid'
}

# New function to select important features using Random Forest
def select_important_features(target, data, label_encoders, importance_threshold=0.1):
    print(f"\n{'='*50}")
    print(f"Performing feature selection for: {target}")
    print(f"{'='*50}")

    # Start timing
    start_time = time.time()

    # Get all numerical features from the dataset (excluding target variables)
    all_targets = list(target_variables.keys())
    numerical_features = [col for col in data.columns if col not in all_targets]

    print(f"Analyzing all {len(numerical_features)} numerical features")

    # Prepare data with all numerical features
    X = data[numerical_features].values

    # Handle missing values if any
    if np.isnan(X).any():
        print("Warning: Dataset contains missing values. Filling with zeros.")
        X = np.nan_to_num(X, nan=0.0)

    # Transform target variable to numeric using the label encoder
    y_encoded = label_encoders[target].transform(data[target].values)

    # Initialize and fit Random Forest model
    rf_model = RandomForestClassifier(**rf_feature_selection_params)
    rf_model.fit(X, y_encoded)

    # Get feature importances
    importances = rf_model.feature_importances_

    # Create DataFrame with feature names and their importances
    feature_importance_df = pd.DataFrame({
        'Feature': numerical_features,
        'Importance': importances
    }).sort_values(by='Importance', ascending=False)

    # Save feature importance data
    feature_importance_df.to_csv(os.path.join(feature_dir, f"{target}_feature_importance.csv"), index=False)

    # Plot feature importances
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(20))
    plt.title(f'Top 20 Feature Importances for {target}')
    plt.tight_layout()
    plt.savefig(os.path.join(feature_dir, f"{target}_feature_importance.png"))

    # Select features above the threshold
    selected_features = feature_importance_df[feature_importance_df['Importance'] >= importance_threshold]['Feature'].tolist()

    print(f"Selected {len(selected_features)} important features (importance >= {importance_threshold}):")
    for feature, importance in zip(feature_importance_df['Feature'][feature_importance_df['Importance'] >= importance_threshold],
                                  feature_importance_df['Importance'][feature_importance_df['Importance'] >= importance_threshold]):
        print(f"  - {feature}: {importance:.4f}")

    # Calculate and print feature selection time
    selection_time = time.time() - start_time
    print(f"Feature selection completed in {selection_time:.2f} seconds")

    return selected_features

# Modified function to train an SVM model with selected features
def train_svm_with_selected_features(target, selected_features, data, label_encoders, save_dir):
    print(f"\n{'='*50}")
    print(f"Training SVM model for: {target}")
    print(f"Using {len(selected_features)} selected features")
    print(f"{'='*50}")

    # Start timing
    start_time = time.time()

    # Verify that all requested features exist in the dataset
    missing_features = [feature for feature in selected_features if feature not in data.columns]
    if missing_features:
        print(f"ERROR: The following features are missing from the dataset: {missing_features}")
        print("Available columns:", data.columns.tolist())
        return

    # If no important features were found, use the original feature mapping as fallback
    if not selected_features and target in original_feature_mapping:
        print(f"No features above threshold were found. Using original feature mapping for {target}.")
        selected_features = original_feature_mapping[target]

    # Prepare data with only the selected features
    X = data[selected_features].values

    # Handle missing values if any
    if np.isnan(X).any():
        print("Warning: Dataset contains missing values. Filling with zeros.")
        X = np.nan_to_num(X, nan=0.0)

    # Feature scaling for SVM (important!)
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    # Save scaler for future use
    scaler_path = os.path.join(save_dir, f"{target}_svm_scaler.pkl")
    with open(scaler_path, 'wb') as f:
        pickle.dump(scaler, f)

    # Transform target variable to numeric using the label encoder
    y_encoded = label_encoders[target].transform(data[target].values)
    num_classes = len(target_variables[target])

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_encoded, test_size=validation_split, random_state=42, stratify=y_encoded
    )

    print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
    print(f"Number of classes: {num_classes}")
    print(f"Feature input shape: {X_train.shape[1]}")

    # Create and train model using GridSearchCV for hyperparameter tuning
    print(f"Performing grid search for optimal SVM hyperparameters...")

    # Initialize the model with grid search
    grid_search = GridSearchCV(
        SVC(),
        svm_param_grid,
        cv=5,
        n_jobs=-1,
        verbose=1
    )

    # Fit model
    grid_search.fit(X_train, y_train)

    # Get best model
    best_model = grid_search.best_estimator_
    print(f"Best parameters: {grid_search.best_params_}")

    # Make predictions on test set
    y_pred = best_model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Test accuracy: {accuracy:.4f}")

    # Print detailed classification report
    print("\nClassification Report:")
    class_names = target_variables[target]
    print(classification_report(y_test, y_pred, target_names=class_names))

    # Create confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names,
                yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix - {target} (SVM with Selected Features)')
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"{target}_svm_confusion_matrix.png"))

    # Save the trained model
    model_path = os.path.join(save_dir, f"{target}_svm_model.pkl")
    with open(model_path, 'wb') as f:
        pickle.dump(best_model, f)

    # Save label encoder
    encoder_path = os.path.join(save_dir, f"{target}_svm_encoder.pkl")
    with open(encoder_path, 'wb') as f:
        pickle.dump(label_encoders[target], f)

    # Save a file with the features used
    features_path = os.path.join(save_dir, f"{target}_svm_features.txt")
    with open(features_path, 'w') as f:
        f.write("\n".join(selected_features))

    # Save the best parameters
    params_path = os.path.join(save_dir, f"{target}_svm_best_params.txt")
    with open(params_path, 'w') as f:
        for param, value in grid_search.best_params_.items():
            f.write(f"{param}: {value}\n")

    # Calculate and print training time
    training_time = time.time() - start_time
    print(f"SVM training completed in {training_time:.2f} seconds")

    return best_model

# Main function to run the feature selection and SVM training pipeline
def run_feature_selection_and_training_pipeline(targets, data, label_encoders, save_dir, importance_threshold=0.1):
    print(f"\n{'#'*80}")
    print(f"# Running Feature Selection and SVM Training Pipeline")
    print(f"# Feature importance threshold: {importance_threshold}")
    print(f"{'#'*80}\n")

    results = {}

    for target in targets:
        print(f"\n\n{'#'*80}")
        print(f"# Processing target: {target}")
        print(f"{'#'*80}\n")

        # Step 1: Feature Selection using Random Forest
        selected_features = select_important_features(
            target=target,
            data=data,
            label_encoders=label_encoders,
            importance_threshold=importance_threshold
        )

        # Step 2: Train SVM model with selected features
        svm_model = train_svm_with_selected_features(
            target=target,
            selected_features=selected_features,
            data=data,
            label_encoders=label_encoders,
            save_dir=save_dir
        )

        # Store results
        results[target] = {
            'selected_features': selected_features,
            'model': svm_model
        }

    return results

# Run the feature selection and training pipeline for all targets
importance_threshold = 0.04  # Features with importance >= 10%
target_list = list(target_variables.keys())

print("\n\n" + "#"*100)
print("# Starting Feature Selection and SVM Training Pipeline")
print("# This process will first use Random Forest to identify important features,")
print(f"# then train SVM models using only features with importance >= {importance_threshold}")
print("#"*100 + "\n")

results = run_feature_selection_and_training_pipeline(
    targets=target_list,
    data=data,
    label_encoders=label_encoders,
    save_dir=save_dir,
    importance_threshold=importance_threshold
)

print("\n" + "#"*100)
print("# Pipeline Complete!")
print("# All models and feature analyses have been saved to:", save_dir)
print("#"*100)
```
